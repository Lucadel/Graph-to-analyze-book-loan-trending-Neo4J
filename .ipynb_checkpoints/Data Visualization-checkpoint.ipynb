```python
import pandas as pd
data_treated=pd.read_csv('data_treated.csv')
```

```python
data_treated.head()
```

```python
from matplotlib import pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,5))
ax1.hist(data_treated['Nombre de prêts 2017'],bins=1000)
ax1.set_xlim(0, 50)
ax2.hist(data_treated['Nombre de prêts 2017'],bins=70)
ax2.set_xlim(51,1500)
ax2.set_ylim(0,10000)
plt.show()
```

```python
import seaborn as sns
import numpy as np
Pret_cat=data_treated.groupby('Catégorie')['Nombre de prêts 2017'].agg(np.sum).sort_values(ascending=False)
plt.figure(figsize=(30,5))
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_cat.index, Pret_cat.values)
plt.xticks(rotation='65', fontsize=19)
plt.title("Nombre de prêts par catégorie",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
chart=sns.set(style="whitegrid")
ax = sns.scatterplot('Nombre d\'exemplaires', 'Nombre de prêts 2017',data=data_treated)
plt.xticks(rotation='0', fontsize=19)
plt.title("Nombre de prêts en fonction du nombre d'exemplaires",fontsize=30)
plt.show()
```

On observe deux types d'ouvrages. Une première catégorie d'ouvrage

```python
Pret_catbis=data_treated.groupby('Type de document')['Nombre de prêts 2017'].agg(np.sum).sort_values(ascending=False)
plt.figure(figsize=(30,5))
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_catbis.index, Pret_catbis.values)
plt.xticks(rotation='65', fontsize=19)
plt.title("Nombre de prêts par type de document",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
Pret_cat2=data_treated.groupby('Catégorie')['Nombre d\'exemplaires'].agg(np.sum).sort_values(ascending=False)
Pret_cat3=(Pret_cat/Pret_cat2).sort_values(ascending=False)
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_cat3.index, Pret_cat3.values)
plt.xticks(rotation='65', fontsize=19)
plt.title("Prêt moyen par exemplaire par catégorie",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
Pret_cat_Musique=data_treated[data_treated['Catégorie']=='Musique'].groupby('Catégorie statistique 1').agg(np.sum)
Pret_cat_Musique2=(Pret_cat_Musique['Nombre de prêts 2017']/Pret_cat_Musique['Nombre d\'exemplaires']).sort_values(ascending=False)
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_cat_Musique2.index, Pret_cat_Musique2.values)
plt.xticks(rotation='90', fontsize=19)
plt.title("Prêt moyen par exemplaire Cat: Musique",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
Pret_cat_Litter=data_treated[data_treated['Catégorie']=='Litterature'].groupby('Catégorie statistique 1').agg(np.sum)
Pret_cat_Litter2=(Pret_cat_Litter['Nombre de prêts 2017']/Pret_cat_Litter['Nombre d\'exemplaires']).sort_values(ascending=False)
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_cat_Litter2.index, Pret_cat_Litter2.values)
plt.xticks(rotation='90', fontsize=19)
plt.title("Prêt moyen par exemplaire Cat: Litterature",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,10))
chart=sns.set(style="whitegrid")
order=Pret_cat3.index
ax = sns.boxplot(x="Catégorie", y="Nombre de prêts par exemplaire", data=data_treated, order=order)
plt.xticks(rotation='65', fontsize=19)
plt.title("Distribution du nombre de prêts par exemplaire par catégorie",fontsize=30)
plt.show()
```
