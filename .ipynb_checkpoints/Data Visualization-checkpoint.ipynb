```python
import pandas as pd
data_treated=pd.read_csv(r'C:\Users\bigdata\_Projet Neo4J\data_treated.csv')
```

```python
import seaborn as sns
import numpy as np
Pret_cat=data_treated.groupby('Catégorie')['prets'].agg(np.sum).sort_values(ascending=False)
plt.figure(figsize=(30,5))
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_cat.index, Pret_cat.values)
plt.xticks(rotation='65', fontsize=19)
plt.title("Nombre de prêts par catégorie",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
Pret_cat2=data_treated.groupby('Catégorie')['Nombre d\'exemplaires'].agg(np.sum).sort_values(ascending=False)
Pret_cat3=(Pret_cat/Pret_cat2).sort_values(ascending=False)
chart=sns.set(style="whitegrid")
ax = sns.barplot(Pret_cat3.index, Pret_cat3.values)
plt.xticks(rotation='65', fontsize=19)
plt.title("Prêt moyen par exemplaire par catégorie",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
chart=sns.set(style="whitegrid")
ax = sns.scatterplot('exemplaires', 'prets',data=data_treated)
plt.xticks(rotation='0', fontsize=19)
plt.title("Nombre de prêts en fonction du nombre d'exemplaires",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(30,5))
chart=sns.set(style="whitegrid")
ax = sns.scatterplot('exemplaires', 'Nombre de prêts par exemplaire',data=data_treated)
ax2=plt.axhline(7, color="red")
ax3=plt.axvline(55, color="red")
ax4=plt.axhline(14, color="red")
plt.xticks(rotation='0', fontsize=19)
plt.title("Nombre de prêts par exemplaire en fonction du nombre d'exemplaires",fontsize=30)
plt.show()
```

A travers les deux pentes distinctes que l'on observe, on identifie plusieurs catégories d'ouvrage. Une première catégorie d'ouvrage dont l'emprunt est en "flux-tendu", c'est  à dire que si le nombre d'exemplaire augmente, le nombre de prets augmenteraient aussi. Une deuxième catégorie à l'inverse "distressed", le nombre de prets est décorrélé du nombre d'exemplaires. Il y a une catégorie de livres avec un fort taux d'emprunt et un faible nombre d'exemplaires disponibles.
